### Problem Statement

This analysis will provide valuable insights into the calorie count and nutrition facts of various menu items.

```python
### CATEGORY = Category of the menu items
### Items = Items name displayed on menu
#### Serving Size = serving size of an item
### Calories = Calories count of an item
### Calories from fat = Calories count from the fat of an item
### Total Fat = Total Fat of an item
### Total Fat(% Daily Value) = Daily percentage value of Fat
### Satuarated Fat = Total Satuarated Fat in an item
### Satuarated Fat(% Daily Value) = Daily percentage value of Satuarated Fat
### Trans Fat = Total Trans Fat of an item
### Choloestrol = Total Cholestrol value in an item
### Choloestrol(% Daily Value) = Daily percentage value of Cholestrol of an item
### Sodium = Total Sodium count in an item
### Sodium(% Daily Value) = Daily percentage value of Sodium of an item
### Carbohydrate = Total Carbohydrates count in an item
### Carbohydrate(% Daily Value) = Daily percentage value of Carbohydrate of an item
### Dietar = Total Dietary fiber in an menu item
### Sugars = Totsl Sugars count in an item
### Protein = Total Protein in the menu Item
### Vitamin A(% Daily Value) = Daily percentage value of Vitamin A
### Vitamin c(% Daily Value) = Daily percentage value of Vitamin C
### Calcium(% Daily Value) = Daily percentage value of Calcium
### Iron(% Daily Value) = Daily percentage value of Vitamin A
```

### Distribution of different nutritions

### Outlier treatment

```python
col=McD.select_dtypes(exclude='object').columns
Q1 = McD[col].quantile(0.25)
Q3 = McD[col].quantile(0.75)
IQR = Q3 - Q1
McD= McD[~((McD[col] < (Q1 - 1.5*IQR))| (McD[col] > (Q3 + 1.5*IQR))).any(axis=1)]
```
